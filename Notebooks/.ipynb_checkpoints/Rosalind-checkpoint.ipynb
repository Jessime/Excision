# Complementing a Strand of DNA

```python
def rev_complement(seq):
    print seq
    print ''
    comp = {'A':'T', 'G':'C','C':'G','T':'A'}
    new_seq = [comp[b] for b in seq]
    new_seq = ''.join(new_seq[::-1])
    print new_seq
```

# Computing GC Content

```python
def max_gc(fasta):
    max_seq = None
    max_gc = 0
    gc = 0
    len_seq = 0
    
    for line in fasta:
        if line['0'] == '>':
            name = line[:1]
            
            
    data = fasta.split('\n')

    for line in data:
        print line
        if line[0] == '>':
            name = line[1:]
        else:
            bases = Counter(line)
            gc = (bases['G'] + bases['C'])/float(len(line.strip()))
            if gc > max_gc:
                max_gc = gc
                max_seq = name
    print max_seq
    print max_gc
```

```python
def max_gc(fasta):
    #build lists
    names_ls = []
    seqs_ls = []
    current_seq = ''
    with open(fasta) as fasta:
        for line in fasta:
            if line[0] == '>':
                names_ls.append(line[1:].strip())
                if current_seq:
                    seqs_ls.append(current_seq)
                    current_seq = ''
            else:
                current_seq += line.strip()
        seqs_ls.append(current_seq)     

    #calc gc
    max_seq = None
    max_gc = 0
    for name, seq in zip(names_ls, seqs_ls):
        gc = (seq.count('G')+seq.count('C'))/float(len(seq))*100
        print name, gc
        if gc > max_gc:
            max_gc = gc
            max_seq = name

    print max_seq
    print max_gc
        
```

# Finding a Motif in DNA

```python
from __future__ import print_function
def occurrences(string, sub):
    c = 0
    while c+len(sub) <= len(string):
        chunk = string[c:c+len(sub)]
        c += 1
        if chunk == sub:
            print(c, end=' ')
```

# Translating RNA into Protein

```python
table = """UUU F      CUU L      AUU I      GUU V
UUC F      CUC L      AUC I      GUC V
UUA L      CUA L      AUA I      GUA V
UUG L      CUG L      AUG M      GUG V
UCU S      CCU P      ACU T      GCU A
UCC S      CCC P      ACC T      GCC A
UCA S      CCA P      ACA T      GCA A
UCG S      CCG P      ACG T      GCG A
UAU Y      CAU H      AAU N      GAU D
UAC Y      CAC H      AAC N      GAC D
UAA Stop   CAA Q      AAA K      GAA E
UAG Stop   CAG Q      AAG K      GAG E
UGU C      CGU R      AGU S      GGU G
UGC C      CGC R      AGC S      GGC G
UGA Stop   CGA R      AGA R      GGA G
UGG W      CGG R      AGG R      GGG G""".split()
table_dic = {k:v for k,v in zip(table[::2], table[1::2])}
```

```python
def translate(dna):
    codons = [dna[i:i+3] for i in xrange(0, len(dna), 3)]
    del codons[-1]
    protein = "".join([table_dic[c] for c in codons])
    return protein
```

# Locating Restriction Sites

```python
def reverse_palandrome(dna):
    size = len(dna)
    compliments = {'A': 'T', 'T':'A', 'C':'G', 'G':'C'}
    for i in xrange(size-4+1):
        for k in xrange(4,13):
            if i+k > size:
                break
            kmer = dna[i:i+k]
            rev_comp = ''.join([compliments[x] for x in kmer][::-1])
            if kmer == rev_comp:
                print i+1, k
```

```python
dna = 'TCAATGCATGCGGGTCTATATGCAT'
reverse_palandrome(dna)
```

# Rabbits and Recurrence Relations

```python
def Fib(months, litter):
    a,b = 1,1
    yield a
    yield b
    for i in xrange(months-2):
        a, b = b, litter*a + b
        yield b
```

```python
for a in Fib(36, 3):
    print a
```

# Introduction to Mendelian Inheritance

```python
def dom_prob(k, m, n):
    s = sum((k,m,n))
    prob = 0
    prob += (k/s)*((k-1)/(s-1))
    prob += (k/s)*(m/(s-1))
    prob += (k/s)*(n/(s-1))
    
    prob += (m/s)*(k/(s-1))
    prob += (m/s)*((m-1)/(s-1))*.75
    prob += (m/s)*(n/(s-1))*.5
    
    prob += (n/s)*(k/(s-1))
    prob += (n/s)*(m/(s-1))*.5
    return prob
```

```python
dom_prob(2., 2., 2.)
```

```python
dom_prob(30., 29., 25.)
```
