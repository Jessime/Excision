I'm going to solve the shortest path across a graph. With networkx, it shouldn't be that difficult.

```python
import numpy as np
```

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

I'll generate a matrix that represents how close 10 cities are to each other as a small example.

```python
dist = np.random.randint(1,50, [10,10])
dist = (dist + dist.T)/2
np.fill_diagonal(dist, 0)
dist[dist<25] = 0
dist
```

```python
G = nx.from_numpy_matrix(dist)
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [12.0, 9.0] #Larger than default of 6, 4
```

```python
pos=nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G,pos,node_size=700)
nx.draw_networkx_edges(G,pos,width=6)
nx.draw_networkx_labels(G,pos,font_size=20,font_family='sans-serif')
plt.axis('off')
plt.show() # display
```

The layouts are weird. I'd have to make the visual somewhere else.

```python
print 'Route:', nx.shortest_path(G,source=6,target=4,weight='weight')
print 'Distance:', nx.shortest_path_length(G,source=6,target=4,weight='weight')
```
