```python
import pandas as pd
```

```python
df = pd.read_csv('pro_model_sig.txt', sep='\t')
df
```

```python
phenotypes = ['tumor', 'protein synthesis', 'lipid synthesis', 'growth', 'cell cycle', 'cytoskeletal activity', 'apoptosis']
```

```python
import numpy as np
```

```python
scores = np.random.random((len(df), len(phenotypes)))
print(scores.shape)
scores
```

```python
negatives = (np.random.randint(0, 2, scores.shape))
print(negatives.shape)
negatives
```

```python
scores[np.where(negatives)] *= -1
```

```python
scores = pd.DataFrame(scores)
```

```python
scores.columns = phenotypes
scores.head()
```

```python
new_df = df.join(scores)
new_df.head()
```

```python
new_df.to_csv('locus_data.txt', sep='\t')
```
